```python
import matplotlib.pyplot as plt

class Feld:
    def __init__(self, xdim, ydim):
        #self.xdim = xdim
        #self.ydim = ydim
        self.X, self.Y = np.meshgrid(np.arange(xdim),np.arange(ydim))
        self.eta = np.zeros_like(self.X)
        self.name = 'Das ist der Name'
    
    def set_func_gauss(self,amplitude = 10,std = 30,position = [10,10]):
        self.eta = amplitude * np.exp(-1/2 *( ((self.X-position[0])/std)**2+((self.Y-position[1])/std)**2))
    
    def plot_initial(self):
        plt.contourf(self.eta)




```

```python
test_feld = Feld(20,20)
test_feld.set_func_gauss(std=2,position=[10,10])
test_feld.plot_initial()
```
